```python
#Usando os componentes do plotly e do pandas, com o uso da biblioteca Dash, para criar gráficos
#Faça uma pequena aplicação Web Dash que exiba um gráfico de barras, usando o componente Graph
#Resposta:
import pandas as pd
import plotly.express as px
from dash import Dash, html, dcc


df = pd.read_csv("dados_aula5.csv")

fig = px.bar(df, x="alunos", y="computadores", title="Computadores por Aluno")

app = Dash()

app.layout = html.Div([
    html.H1("Monitoramento de Computadores por Aluno"),
    dcc.Graph(id="grafico-computadores", figure=fig)
])

if __name__ == "__main__":
    app.run(debug=True)
```

```python
#Usando os componentes do plotly, da biblioteca Dash, para criar gráficos
#Faça uma pequena aplicação Web Dash que exiba um gráfico de barras, usando o componente Graph
#Que seja capaz de ler um arquivo CSV, criado por voce !
#E manipule o grafico com os diferntes tipos de gráficos, como linha, pizza, etc... usando o plotly express

#Resposta:

import pandas as pd
import plotly.express as px 
from dash import Dash, dcc, html 

df = pd.read_csv("dados_aula5.csv") 

fig = px.bar(
    df,
    x="alunos",
    y="computadores",
    title="Computadores por Aluno",
    labels={
        "alunos": "Alunos",
        "computadores": "Computadores"
    }
)
app = Dash()
app.layout = html.Div(
    children=[
        html.H1("Monitoramento de Computadores por Aluno"),
        dcc.Graph(
            id="grafico-computadores",
            figure=fig
        )
    ]
)
if __name__ == "__main__":
    app.run(debug=True)
```
