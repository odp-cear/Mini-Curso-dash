```python
#Sabendo dos conceitos basicos sobre callbacks, e pondo seus conhecimentos em pratica,
#crie um codigo com o uso do callback que atualize o grafico de acordo com o valor do dropdown.
#Se possivel use os componentes input e output do dash, para uma melhor conexão entre o codigo e os dados lidos.

#Resposta:

from dash import Dash, dcc, html, Input, Output

app = Dash(__name__)

app.layout = html.Div([
    dcc.Dropdown(
        id="meu_dropdown",
        options=[
            {"label": "Opção 1", "value": "Você escolheu a opção 1"},
            {"label": "Opção 2", "value": "Você escolheu a opção 2"},
        ],
        value="Você escolheu a opção 1"
    ),
    html.Div(id="meu_texto")
])

@app.callback(
    Output("meu_texto", "children"),
    Input("meu_dropdown", "value")
)
def atualizar_texto(valor):
    return valor

if __name__ == "__main__":
    app.run_server(debug=True)

```
