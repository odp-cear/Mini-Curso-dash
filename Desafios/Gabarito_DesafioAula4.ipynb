```python
# Agora Usando os componentes Html, da biblioteca Dash, para criar a estrutura da página
# Faça uma pequena aplicação Web Dash que exiba um Div que contenha uma frase de exibição.
# E dentro do Div, adicione um componente para Titulo, um componente para Parágrafo!

#Resposta:

from dash import Dash, html

app = Dash()
app.layout = html.Div(children=[
    html.H1('Usando o componente H1 como um título!'),
    html.P('Usando o componente P como um parágrafo!')
])

if __name__ == '__main__':
    app.run(debug=True)

```
