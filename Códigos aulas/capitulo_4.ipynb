1‚Å∞ ponto - Componentes do Dash HTML.

No Dash,Inumeras funcionalidades apresentadas como conteudo do dashbord √© importada para HTML ,uma das bibliotecas principais e mais usadas √© a biblioteca dash_html_components (importada geralmente como html). Cada tag HTML nativa possui um componente correspondente em Python.

A l√≥gica: O nome da classe no Python √© id√™ntico √† tag HTML, mas com a primeira letra mai√∫scula (ex: html.H1, html.Div, html.P).

O argumento children: Quase todos os componentes aceitam um argumento chamado children, que √© onde voc√™ coloca o conte√∫do (texto ou outros componentes) que ficar√° dentro daquela tag.

2‚Å∞ ponto - Divs,T√≠tulos, paragrafos e elementos estruturais

Para que o nosso c√≥digo n√£o seja apenas um conjunto de informa√ß√µes de texto, usamos os pilares da organiza√ß√£o web:

- html.Div: √â o "container" gen√©rico. Serve para agrupar outros elementos e aplicar estilos a blocos inteiros. No Dash, voc√™ usar√° Divs constantemente para criar se√ß√µes e colunas.

- html.H1 a html.H6: S√£o os t√≠tulos. O H1 √© o principal da p√°gina, enquanto os outros definem a hierarquia de import√¢ncia do conte√∫do.

- html.P: O par√°grafo. Utilizado para blocos de texto corrido, garantindo o espa√ßamento correto entre blocos de informa√ß√£o.

Men√ß√µes honrosas para os componentes Html.label e Html.table,componentes tambem bastante usados para criar componentes interativos e tabelas.

Veja a estrutura e a ordem de uso dos elementos acima citados, no codigo abaixo:

```python
#Exemplo usado na nossa aula 4:
from dash import Dash, html

# Criando a aplica√ß√£o Dash
app = Dash()
# Layout da p√°gina
app.layout = html.Div(className='container', children=[
#html.Div √© respons√°vel por criar um container de informa√ß√µes que sera exibido na p√°gina!
# Bloco 1: Div contendo H1 e P

                html.H1('Aula 4 Estrutura b√°sica com Dash html!',className='titulo'),
#html.H1 √© respons√°vel por criar um t√≠tulo ou cabe√ßalho na p√°gina!
                html.P('Este bloco demonstra o uso de t√≠tulos e par√°grafos dentro de uma Div no Dash.',className='paragrafo'),
                html.Div(className='paragrafo1', children=['Tambem podemos usar a Div como sendo um paragrafo!.'])
            ]#html.P √© respons√°vel por criar um par√°grafo na p√°gina!
        ),

# Bloco 2: Div simples usada como par√°grafo
html.Div(className='paragrafo1', children=['Tambem podemos usar a Div como sendo um paragrafo!.'])

if __name__ == '__main__':
    app.run(debug=True)
```

3‚Å∞ ponto - Montando uma P√°gina Simples

A cria√ß√£o de um layout organizado no Dash segue uma estrutura de √°rvore. O app.layout recebe um componente principal (geralmente uma html.Div) que cont√©m uma lista de todos os outros elementos.

Um layout bem organizado deve seguir esta ordem l√≥gica:

Header: T√≠tulo e talvez um subt√≠tulo (html.H1/P).

Body: Divs principais que separam os gr√°ficos dos controles.

4‚Å∞ ponto - Estilos B√°sicos (Inline vs. Externo)

Para deixar o layout visualmente agrad√°vel, voc√™ tem dois caminhos principais:

Estilos Inline: Voc√™ passa um dicion√°rio Python para o argumento style.

Exemplo: style={'color': 'blue', 'fontSize': '20px'}. Note que no Dash usamos chaves para o (fontSize) em vez do h√≠fen do CSS tradicional (font-size).

CSS Externo: √â a forma mais limpa. Voc√™ cria uma pasta chamada assets no mesmo diret√≥rio do seu script .py e coloca um arquivo .css l√° dentro. O Dash carrega esse arquivo automaticamente, permitindo que voc√™ use o argumento className nos componentes para aplicar classes CSS.

Veja abaixo o codigo inline usado na video aula 4 , e em seguida o codigo com o estilo usado via CSS externo:

```python
#Codigo referente ao uso da customiza√ß√£o inline:
from dash import Dash, html

# Criando a aplica√ß√£o Dash
app = Dash()
div_style = {
    'fontFamily': 'Sans-serif',
    'fontSize': '18px',
    'textAlign': 'center',
    'padding': '20px',
    'color': "#000000",
}
# Layout da p√°gina
app.layout = html.Div( style=div_style, className='container', children=[
#html.Div √© respons√°vel por criar um container de informa√ß√µes que sera exibido na p√°gina!
# Bloco 1: Div contendo H1 e P

                html.H1('Aula 4 Estrutura b√°sica com Dash html!',className='titulo', style=div_style),
#html.H1 √© respons√°vel por criar um t√≠tulo ou cabe√ßalho na p√°gina!
                html.P('Este bloco demonstra o uso de t√≠tulos e par√°grafos dentro de uma Div no Dash.',)
            ]#html.P √© respons√°vel por criar um par√°grafo na p√°gina!
        ),

# Bloco 2: Div simples usada como par√°grafo
html.Div(children=['Tambem podemos usar a Div como sendo um paragrafo!.'])

if __name__ == '__main__':
    app.run(debug=True)
```
