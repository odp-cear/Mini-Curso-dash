Ponto 2 - O que é um diretório?

- Um diretório é o local onde você está programando,
salvando suas alterações, arquivos e dados.
- Um diretório é a pasta na qual o programador está
guardando os seus arquivos
- Um diretório é o caminho para onde os arquivos e alterações vão.

-> Como prosseguir no curso Dash?
- Abrir uma pasta vazia;
- Abrir essa pasta criada no VScode;
- Clonar repositório do Github no Vscode;
- Salvar os arquivos do repositório na pasta criada.

Porque é importante estar no diretório correto?
- Para que o programador tenha um melhor controle e organização;
- Para conseguir efetuar alterações;
- Para usar todos os dados, arquivos e bases que são necessárias para gerar seus dashboards
|-> precisam estar no mesmo diretório!!!
