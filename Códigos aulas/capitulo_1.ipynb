Ex pr√°tico de um dashboard criado com Dash:

```python
#C√≥digo utilizado de exemplo na aula 1:
from dash import Dash, html, dcc
import plotly.express as px
import pandas as pd

#Criando o app dash
app = Dash()

#Assumir a exist√™ncia de um Dataframe
df = pd.DataFrame({
    "Fruta": ["Ma√ß√£s", "Laranjas", "Bananas","Ma√ß√£s", "Laranjas", "Bananas"],
    "Quantidade": [4,1,2,2,4,5],
    "Cidade": ["S√£o Paulo", "S√£o Paulo", "S√£o Paulo", "RJ", "RJ", "RJ" ]
})

fig = px.bar(df, x="Fruta", y="Quantidade", color="Cidade",)

#Construindo o desenho do dashboard
app.layout = html.Div(children=[
    html.H1(children='An√°lise de frutas'),

    html.Div(children='''
        Dash: uma estrutura de aplica√ß√£o na web para os seus dados.
    '''),
    
    dcc.Graph(
        id='example-graph',
        figure=fig
    )
])

if __name__ == '__main__':
    app.run(debug=True) #rodar o app
```
